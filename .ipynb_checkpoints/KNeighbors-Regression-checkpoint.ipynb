```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split

from sklearn.neighbors import KNeighborsRegressor as KNN
from sklearn.metrics import mean_squared_error as mse
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
data = pd.read_csv("train_cleaned.csv")
data
```

```python
data.head()
```

```python
x=data.drop(['Item_Outlet_Sales'],axis=1)
y=data['Item_Outlet_Sales']
```

```python
x.shape , y.shape
```

```python
mms = MinMaxScaler()
```

```python
x_scaled=mms.fit_transform(x)
```

```python
x=pd.DataFrame(x_scaled)
x
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=56)
```

```python
x_train.shape , x_test.shape , y_train.shape , y_test.shape
```

```python
model=KNN(n_neighbors=5)
```

```python
model.fit(x_train,y_train)
```

```python
test_predict=model.predict(x_test)
test_predict
```

```python
k=mse(test_predict,y_test)
```

```python
k
```

# Elbow Method 

```python
def elbow(K):
    test_error=[]
    
    for i in K:
        model=KNN(n_neighbors=i)
        model.fit(x_train,y_train)
        test_predict=model.predict(x_test)
        error=mse(test_predict,y_test)
        test_error.append(error)
    return test_error
```

```python
k=range(1,40)
test=elbow(k)
```

```python
test
```

```python
plt.plot(k,test)
plt.xlabel('K neighbor')
plt.ylabel('Test Mean Square Error')
plt.title("Elboe Curve for test")
```
